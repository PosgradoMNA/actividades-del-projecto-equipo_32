**2. Escribe los pasos estándares en un máximo de 3 oraciones, por cada fase**

**Comprensión del negocio**

Se determinan los objetivos generales, se hace un análisis de la situación para recolectar los requisitos del proyecto desde una perspectiva de negocio, una vez que se determina el alcance, se elabora un plan de actividades.

**Comprensión de los datos**

Se realiza la recolección inicial de los datos para poder establecer un primer contacto con el problema, se averigua su calidad, también se busca de manera anticipada relaciones, tendencias para obtener mayor comprensión de los datos, así como identificar las relaciones más evidentes.

**Preparación de los datos**

Se determinan los datos que se utilizarán para la construcción del modelo, es posible la integración de datos adicionales de múltiples fuentes. Esto implica la limpieza de los datos, así como la modificación de estos para la selección de las variables más significativas.

**Modelo**

Se selecciona la técnica de modelado del análisis a utilizar, se generan pruebas de diseño para su construcción. Generalmente se evalúan múltiples modelos que someten a un régimen de pruebas a partir de muestras que tengan representatividad de la población de estudio.

**Evaluación**

Se evalúa la utilidad de los resultados de los modelos por parte de los expertos del tema, se revisan los procesos logrados y se determinan los siguientes pasos para determinar si se procede con el despliegue e implementación del modelo.
 
**Despliegue**

Se desarrolla un plan de despliegue del modelo, incluye el mantenimiento y monitoreo. Se entrega la documentación con la arquitectura del proyecto, se especifica el proceso de la minería de datos, así como el alcance y resultados.
