# Parte 1: Fundamentos de bases de datos
**Fundamentos de bases de datos y almacenes de datos (Data Warehouse) para ciencia de datos.**

Una base de datos se puede comprender como un conjunto de listas llamadas tablas compuestas de columnas,filas y celdas la cual recopila de manera compartida datos logicamente relacionados con el objetivo de facilitar la consulta del usuario y cumplir con los objetivos fijados por la organizacion. La base de datos requiere un sistema de administracion de base de datos (DBMS) la cual se utiliza para mantener controlado el acceso a los usuarios de la base de datos, para acceder a los datos e interactuar con ellos de manera adecuada podemos utilizar SQL. Algunas de las ventajas de las bases de datos son:

1.- Mejor calidad de datos

2.- Mejor accesibilidad y calidad de respuesta

3.- Redundancia minima de datos.

Tambien tenemos que tomar en cuenta los riesgos y costos al implementar una base de datos para la organizacion como pueden ser:

1.-Vulnerabilidad

2.-Costo de software y hardware

3.-Complejidad

Una Data Warehouse es una enorme base de datos que es alimentada por multiples fuentes de datos los cuales se unen en esa enorme base de datos. Un ejemplo es como cuando un banco quiere almacenar informacion de una empresa que se dedica a la venta de articulos en linea, hay muchos datos que se pueden obtener y los cuales posteriormente pueden ser consultados atravez de SQL(Lenguaje de consulta estructurado) utilizando ciertas herramientas de analisis de datos como lo es tableau. Tambien debemos saber que al tener una enorme cantidad de datos se tienen que realizar diversas operaciones para eveluar los datos que se estan obteniendo como por ejemplo tenemos que normalizar los datos, evaluar si son comparales entre columnas,datos faltantes o valores atipicos asi como las transformaciones que se requieran realizar

**Uniformidad y Pre-procesamiento de datos**

En la vida real normalmente los datos que se nos presentan no siempre llegaran a la base de datos como nos gustaria que estuvieran en el modelo sino al contrario, normalmente llegan incompletos o inconsistentes para poder procesarlos por lo tanto tenemos que trabajarlos previamente para que sean comprensibles y posteriormente los validamos. La tecnica de preprocesamientos de datos se utiliza para transformar o tratar los datos que no han sido procesados en un formato comprensible y de esa manera los podemos trabajar en el modelo sin crear tendencias o comportamientos que no seran confiables. Para ellos tenemos ciertas tareas que ejecutaremos como los son:

1.- Limpieza de datos.

2.- Transformacion de datos.

3.- Reduccion de datos.
